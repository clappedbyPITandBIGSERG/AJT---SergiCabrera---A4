```python
def fibonacci(n):
    if n==0:
        return[0]
    elif n==1:
     return[0,1]
    else:
       fibonacci =[0,1]
       for i in range (2, n):
        fibonacci.append(fibonacci[i-1]+fibonacci[i-2])
       return fibonacci

n=30
resultat = fibonacci(n)
print(resultat)
```

```python
from IPython.display import clear_output

def mostra_taulell(taulell, diccionari_jugadors):
    print("    " + "   ".join(str(i) for i in range(3)))
    for fila in range(3):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(3))
        print(f" {fila}  {contingut}")
        if fila < 2:
            print("   " + "---+" *2 + "---")
    print()
    return


def demana_posicio(taulell):
    try:
        fila = int(input("Dona'm la fila on vols posar la peça (valor entre 0 i 2)"))
        while fila < 0 or fila > 2:
            fila = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i 2)"))
    except:
        return None, None
    try:
        columna = int(input("Dona'm la columna on vols posar la peça (valor entre 0 i 2)"))
        while columna < 0 or columna > 2:
            columna = int(input("Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i 2)"))
    except:
        return None, None
    return fila, columna

def actualitza_taulell(taulell, fila, columna, jugador):
    taulell[fila][columna] = jugador
    return taulell

def canvi_jugador(jugador):
    return jugador*(-1)

def comprovar_victoria(taulell,jugador):
    # Victòria per files
    for fila in range(0,3):
        suma = 0
        for columna in range(0,3):
            suma += taulell[fila][columna]
        if suma == 3*jugador:
            return True
    # Victòria per columnes
    for columna in range(0,3):
        suma = 0
        for fila in range(0,3):
            suma += taulell[fila][columna]

        if suma == 3*jugador:
            return True
    # Victòria per diagonals
    suma = 0
    for diagonal in range(0,3):
        suma += taulell[diagonal][diagonal]            
    if suma == 3*jugador:
        return True
    # Victòria per diagonal
    suma = 0
    for diagonal in range(0,3):
        suma += taulell[diagonal][2-diagonal]            
    if suma == 3*jugador:
        return True
   
    return False
def taulell_complet(taulell):
    for fila in range(0,3):
        for columna in range(0,3):
            if taulell[fila][columna] == 0:
                return False
    return True

```

```python
jugador = 1
diccionari_jugadors = {
    0: " ",
    1: "X",
    -1: "O"
    }

taulell=[[0, 0, 0], [0, 0, 0], [0, 0, 0]]
condicio_victoria = False
comptador = 0

while not condicio_victoria and comptador < 9:
    clear_output(wait=True)
    mostra_taulell(taulell,diccionari_jugadors)
    fila, columna = demana_posicio()
    if fila == None and columna == None:
        print("Has de donar un nombre natural entre 0 i 2. Sortim del joc")
        break
    taulell = actualitza_taulell(taulell, fila, columna, jugador)
    condicio_victoria = comprovar_victoria(taulell, jugador)
    if condicio_victoria:
        clear_output(wait=True)
        mostra_taulell(taulell,diccionari_jugadors)
        print(f"Ha guanyat el jugador {diccionari_jugadors[jugador]}")
        break

    comptador += 1
    jugador = canvi_jugador(jugador)

if taulell[columna][fila] == 0:
   print ("casella desocupada")
else:
 print("casella ocupada")
```
